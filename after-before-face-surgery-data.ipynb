```python
import pandas as pd
import cv2
import numpy as np

# Initialize the dictionary
dataset = {}
image_path_list = []
# Assume `image_path_list` is a list of image file paths
for i in range(1,93):
    image_path = f"/kaggle/input/plastic-surgery-face/plastic face/blepharoplasty-eyelid-surgery-92/after/{i}.jpg"
    image_path_list.append(str(image_path))

for i, image_path in enumerate(image_path_list, start=1):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Resize the image to 92x112 pixels
    image_resized = cv2.resize(image, (92, 112))

    # Flatten the image to a 1D array
    pixel_values_1d = image_resized.flatten()

    # Assign the 1D pixel matrix to the dictionary with the current index `i` as the key
    dataset[i] = pixel_values_1d


with open('dataset.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    
    # Write header (optional, e.g., "Key", "Pixel1", "Pixel2", ...)
    header = [f"Pixel{i+1}" for i in range(len(next(iter(dataset.values()))))]
    writer.writerow(header)
    
    # Write each key and its associated list as a row in the CSV
    for key, pixel_values in dataset.items():
        writer.writerow([key] + pixel_values)



print("New column added and saved to 'data_modified.csv'")


print("Dictionary has been successfully written to 'dataset.csv'")

df = pd.read_csv('dataset.csv')

# Step 2: Add a new column with incrementing values
df['Label'] = range(1, 93)  # Adjust range if more or fewer rows are expected

# Step 3: Save the modified DataFrame back to a CSV file
df.to_csv('dataset.csv', index=False)
df = pd.read_csv('dataset.csv')
print(df.head())
```
